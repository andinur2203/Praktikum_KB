```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Penyebab Kematian di Indonesia yang Dilaporkan.csv")
df.info()
```

# 1. Grafik Pie
## Berdasarkan Dataset di atas dengan menggunakan Grafik Pie dapat disimpulkan bahwa: 
## Penyebab kematian terbesar di Indonesia adalah Bencana Non Alam dan Penyakit dengan persentase 75.3%, diikuti oleh Bencana Alam sebesar 21.8% dan Bencana Sosial sebesar 2.9%

```python
genres = df.dropna(subset=['Type']).sort_values(by=['Type'])['Type'].unique()

total = df.groupby(['Type'])['Type'].count()

patches, texts, pcts = plt.pie(total, labels=genres, autopct='%.1f%%', wedgeprops={'linewidth': 1.0, 'edgecolor': 'white'}, textprops={'size': 'x-large'})

plt.title('Diagram Penyebab Kematian Terbesar')
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```

# 2. Grafik Batang
## Berdasarkan Dataset di atas dengan menggunakan Grafik Batang dapat disimpulkan bahwa: 
## Total kematian tertinggi terjadi pada tahun 2004 dan 2021.

```python
x = list(df.iloc[:,2])
y = list(df.iloc[:,4])
plt.figure(figsize = (10,5))
plt.bar(x, y, color = "blue")
plt.title("Perbanding Total Kematian Per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Total Kematian")
plt.show()
```

# 3. Scatter Plot
## Berdasarkan Dataset di atas dengan menggunakan Scatter Plot dapat disimpulkan bahwa: 
## Kasus kematian terbanyak terjadi pada rentang tahun 2005 - 2010.

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10,7))
sns.scatterplot(
data = df,
hue = 'Type',
x = 'Year',  
y = 'Total Deaths')
plt.title('Persebaran Kasus Kematian Per Tahun')
plt.show()
```
